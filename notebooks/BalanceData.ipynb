```python
import random, os, shutil, math
destpath = 'splitDataset'       #directory to host new dataset, will contain two folder 'train' and 'validation' each one with all 14 classes in a specific subfolfer
origpath='training'      #directory of original dataset
classes=['Apple','Blueberry','Cherry','Corn','Grape','Orange','Peach','Pepper','Potato','Raspberry','Soybean','Squash','Strawberry','Tomato']
othersList=['Apple','Blueberry','Cherry','Corn','Grape','Orange','Peach','Pepper','Potato','Raspberry','Soybean','Squash','Strawberry']
```

```python
#split original dataset in two sets for train and validation, assigning 80% to training, and 20% to validation

for i in classes:
  sourceFolder=origpath  + '/'+i
  trainFolder= destpath+ '/train/' + i
  validFolder = destpath+ '/validation/' + i

  filenames=os.listdir(sourceFolder)
  trainFileNames=random.sample(filenames,math.floor(0.8*len(filenames)))
  validationFileNames=list(set(filenames) - set(trainFileNames))

  for fname in trainFileNames:
    srcpath = os.path.join(sourceFolder, fname)
    shutil.copy(srcpath, trainFolder)

  for fname in validationFileNames:
    srcpath = os.path.join(sourceFolder, fname)
    shutil.copy(srcpath, validFolder)
```

```python
# balance selected folder by oversampling and undersampling:

def balance(subdir):
  
  # consider the second most represented class, and randomly duplicate images in other classes to reach that number

  if subdir == 'train':
    maxClassSize=1398
    tomatoSize= 4554
  else:
    maxClassSize= 350
    tomatoSize=1139

  for i in othersList:
    sourceFolder='splitDataset'+ '/'+ subdir + '/' + i
    filenames=os.listdir(sourceFolder)
    length=len(filenames)
    if length<maxClassSize:
      copies=math.floor(maxClassSize/length)
      for fname in filenames:
        srcpath = os.path.join(sourceFolder, fname)
        for j in range(copies-1):
          shutil.copy(srcpath, sourceFolder+'/' + str(j+1) + 'copia'+fname)
      filenames = random.sample(os.listdir(sourceFolder), maxClassSize-(length*copies))
      for fname in filenames:
        srcpath = os.path.join(sourceFolder, fname)
        shutil.copy(srcpath, sourceFolder+'/' + str(copies) + 'copia'+fname)

  # then simply randomly cut half of the tomatoes pictures

  sourceFolder= 'splitDataset'+ '/'+  subdir+'/Tomato'
  filenames = random.sample(os.listdir(sourceFolder), math.floor(tomatoSize/2))
  for fname in filenames:
    srcpath = os.path.join(sourceFolder, fname)
    os.remove(srcpath)
```

```python
# balance both train and validation folder

sourceFolder='splitDataset'
balance('train')
balance('validation')
```
